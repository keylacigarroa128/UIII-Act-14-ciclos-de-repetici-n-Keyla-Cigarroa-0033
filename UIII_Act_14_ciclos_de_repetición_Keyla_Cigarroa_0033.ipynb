### Problema 12
 - Realizar un algoritmo para determinar cuánto ahorrará una persona en un año, si al final de cada mes deposita cantidades variables de dinero; además, se quiere saber cuánto lleva ahorrado cada mes.



```python
# Inicializamos la variable para acumular el total ahorrado
total_ahorrado = 0

print("--- CONTROL DE AHORRO MENSUAL ---")

# Iteramos mes a mes durante un año (12 meses)
for mes in range(1, 13):
    # Solicitamos la cantidad ahorrada en el mes actual
    print(f"\nMes {mes}:")
    ahorrito_mes = float(input("¿Cuánto dinero vas a depositar este mes?: $"))

    # Validamos que no sea una cantidad negativa usando un if-else básico
    if ahorrito_mes < 0:
        print("Cantidad no válida. Se registrará como $0 para este mes.")
        ahorrito_mes = 0
    else:
        # Si es una cantidad válida, mostramos un mensaje de éxito
        if ahorrito_mes > 0:
            print("¡Excelente depósito!")
        else:
            print("Este mes no hubo depósito.")

    # Sumamos el depósito al acumulador total
    total_ahorrado += ahorrito_mes

    # Mostramos el dinero que lleva acumulado hasta el momento
    print(f"Llevas ahorrado acumulado hasta el mes {mes}: ${total_ahorrado:.2f}")

# Al salir del ciclo, mostramos el gran total del año
print("\n-------------------------------------------")
print(f"¡Fin del año! El total ahorrado en los 12 meses es: ${total_ahorrado:.2f}")
print("Programa Realizado por Keyla Paola Cigarroa Palacios")
print("-------------------------------------------")
```
